```python
df = pd.read_csv("/Users/esther/Desktop/Columbia-Research-Project/Yahoo Finance/WU2.csv")
c2 = df["Surprise (%)"]
c3 = []
c4 = []
c5 = []

for i in c2:
    if i == 0:
        c3.append("None")
    if i < 0:
        c3.append("negative")
    if i > 0:
        c3.append("positive")

     
for i in c2:
    if i == 0:
        c4.append("None")
    if i < 0:
        if abs(i) <= 5:
            c4.append("Neg_Average")
        if 5 < abs(i) <= 10:
            c4.append("Below_Average")
        if abs(i) > 10:
            c4.append("Significantly_Below")
    if i > 0:
        if i <=5:
            c4.append("Pos_Average")
        if 5< i <= 10:
            c4.append("Above_Average")
        if i > 10:
            c4.append("Significantly_Above")


for i in range(len(c4)-2):
    if c2[i] == 0:
        c5.append("NA")
    if c2[i] < 0:
        if c2[i+1]<0 and c2[i+2]<0:
            c5.insert(i,"UnderP")
        else:
            c5.insert(i,"regular")
    if c2[i] > 0:
        if c2[i+1]>0 and c2[i+2]>0:
            c5.insert(i,"OutP")
        else:
            c5.insert(i,"regular")
c5.append("regular")
c5.append("regular")


df["Sign"] = c3
df["Expectation"] = c4
df["Performance"] = c5
df.to_csv("WU3.csv")

        


```
